```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from collections import Counter
import nltk
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score
```

```python
df = pd.read_pickle('data/cleaned_df.pkl')
```

```python
df.head(13)
```

TF-IDF: “Term Frequency times Inverse Document Frequency”

```python
X_body_text = df['clean_text'].values
X_title_text = df['clean_title'].values
y = df['label'].values
```

```python
TF_IDF = TfidfVectorizer(ngram_range=(1,2), max_df= 0.85, min_df= 0.01)
```

```python
X_body_TF_IDF = TF_IDF.fit_transform(X_body_text)
X_title_TF_IDF = TF_IDF.fit_transform(X_title_text)
```

```python
indices = df.index.values
```

```python
X_body_TF_IDF_train, X_body_TF_IDF_test, \
y_body_train, y_body_test, \
indices_body_train, indices_body_test = train_test_split(X_body_TF_IDF, y, indices, test_size = 0.2, random_state=42)
```

```python
df.loc[indices_body_train].groupby('label').agg('count')
```

```python
df.loc[indices_body_test].groupby('label').agg('count')
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb_body = MultinomialNB()
```

```python
nb_body.fit(X_body_TF_IDF_train, y_body_train)
```

```python
y_body_train_pred = nb_body.predict(X_body_TF_IDF_train)
```

```python
print('Naive Bayes In Sample F1 and Accuracy Scores:')
print('F1 score {:.4}%'.format(f1_score(y_body_train, y_body_train_pred, average='macro')*100 ))
print ('Accuracy score {:.4}%'.format(accuracy_score(y_body_train, y_body_train_pred)*100))
```

```python
np.where(y_body_train != y_body_train_pred)
```
